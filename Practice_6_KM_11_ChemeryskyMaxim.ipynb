"William Shakespeare" -> "A small spark"

Перша фраза містить 11 унікальних літер (хоча деякі повторюються): a, e, h, i, k, l, m, p, r, s та w. Друга фраза містить 7 унікальних літер: a, k, l, m, p, r та s. Кожна з літер другої фрази зустрічається принаймні один раз у першій фразі.

## **Практичне зайняття №6**
### Тема: "Робота з кортежами, множинами та словниками"

```
word1 = input("введіть 1 фразу: ")
letters1 = {1}
for i in word1:
    if i.isalpha() == True:
        letters1.add(i.lower())
letters1.discard(1)

word2 = input("введіть 2 фразу: ")
letters2 = {1}
for i in word2:
    if i.isalpha() == True:
        letters2.add(i.lower())
letters2.discard(1)
if letters1 >= letters2:
    poss = "2 фразу з 1 побудувати можна"
else:
    poss = "2 фразу з 1 побудувати не можна"

print(letters1)
print(letters2)
print(poss)
```
